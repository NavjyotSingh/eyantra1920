```python
import keras
import pandas as pd
import numpy as np
import sklearn as sk
import matplotlib.pyplot as plt
import ciso8601
import time
```

```python
df1 = pd.read_csv("D:/AIDE/Data/human-activity-smart-devices/glasses.csv")
df2 = pd.read_csv("D:/AIDE/Data/human-activity-smart-devices/report.csv")
```

```python
df1.head(70)
```

```python
#FOR DF1
from datetime import datetime

INPUT_FORMAT = '%Y/%m/%d %H:%M:%S.%f'
OUTPUT_FORMAT = '%Y%m%d %H%M%S'

df1['DATE'] = df1['DATE'].apply(lambda x: datetime.strptime(x,INPUT_FORMAT).strftime(OUTPUT_FORMAT))
df1['DATE'] = df1['DATE'].apply(lambda x: time.mktime(ciso8601.parse_datetime(x).timetuple()) )
```

```python
df1.head()
```

```python
#FOR DF2


INPUT_FORMAT = '%d-%m-%Y %H:%M'
OUTPUT_FORMAT = '%Y%m%d %H%M%S'

df2['to'] = df2['to'].apply(lambda x: datetime.strptime(x,INPUT_FORMAT).strftime(OUTPUT_FORMAT))
df2['to'] = df2['to'].apply(lambda x: time.mktime(ciso8601.parse_datetime(x).timetuple()) )

df2['from'] = df2['from'].apply(lambda x: datetime.strptime(x,INPUT_FORMAT).strftime(OUTPUT_FORMAT))
df2['from'] = df2['from'].apply(lambda x: time.mktime(ciso8601.parse_datetime(x).timetuple()) )
```

```python
df2.shape
```

```python
df2.rename( columns = {'from' : 'DATE'}, inplace = True)
df1.rename( columns = {'DATE' : 'DATE'}, inplace = True)
```

```python
df3 = pd.merge(df1[['DATE', 'ACC_X','ACC_Y','ACC_Z', 'GYRO_X','GYRO_Y','GYRO_Z']],df2[['DATE','duration','activity_type']],left_on='DATE',right_on='DATE', how = 'left')
```
