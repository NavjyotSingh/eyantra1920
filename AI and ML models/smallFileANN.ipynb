```python
import keras
import pandas as pd
import numpy as np
import sklearn as sk
import matplotlib.pyplot as plt
import ciso8601
import time
```

```python
df1 = pd.read_csv("D:/AIDE/Data/run-or-walk/dataset.csv")
```

```python

df1['timestamp']= df1['date'].map(str)+" "+df1['time']



df1['timestamp'] = df1['timestamp'].apply(lambda x : x[:-3])



from datetime import datetime

INPUT_FORMAT = '%d-%m-%Y %H:%M:%S:%f'
OUTPUT_FORMAT = '%Y%m%d %H%M%S'

df1['timestamp'] = df1['timestamp'].apply(lambda x: datetime.strptime(x,INPUT_FORMAT).strftime(OUTPUT_FORMAT))
df1['timestamp'] = df1['timestamp'].apply(lambda x: time.mktime(ciso8601.parse_datetime(x).timetuple()) )

df1 = df1.drop(['username','wrist','date','time'],axis=1)
```

```python
headers = df1.columns
featuresHeaders=['acceleration_x','acceleration_y','acceleration_z','gyro_x','gyro_y','gyro_z']


targetHeader = headers[0]

features = df1[featuresHeaders]
target = df1[targetHeader]
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn import datasets, linear_model
```

```python
featuresTrain, featuresTest, targetTrain, targetTest = train_test_split(features, target, test_size=0.40, random_state=9)
```

```python
import keras 
from keras import backend as K
from keras.models import Sequential
from keras.layers import Activation
from keras.layers.core import Dense
from keras.optimizers import Adam
from keras.metrics import categorical_crossentropy
```

```python
ip = np.array(featuresTrain)
op = np.array(targetTrain)
ipTest = np.array(featuresTest)
opTest = np.array(targetTest)

model = Sequential([
    Dense(8, input_shape=(6,), activation='relu'),
    Dense(4, activation='relu'),
    Dense(2, activation='softmax')
])
```

```python
model.compile(Adam(lr=.0001), loss='sparse_categorical_crossentropy', metrics=['accuracy'])

```

```python
model.fit(ip, op, batch_size=10,validation_split=0.05, epochs=20, shuffle=True, verbose=2)
```

```python
predictions = model.predict_classes(ipTest, batch_size=1, verbose=5)
```

```python
model.evaluate(featuresTest,targetTest)
```
